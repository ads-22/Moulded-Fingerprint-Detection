```python
from PIL import Image
import csv
import os
data=[]
with open('images.csv', 'w', newline='') as writeFile:
    writer = csv.writer(writeFile)
    data.append("filename")
    data.append("real/altered")
    writer.writerow(data)
    data=[]
    for filename in os.listdir("/content/SOCOFing/Real"):
        data.append(filename)
        data.append("real")
        writer.writerow(data)
        data=[]
    for filename in os.listdir("/content/SOCOFing/Altered/Altered-Easy"):
        data.append(filename)
        data.append("altered")
        writer.writerow(data)
        data=[]
    for filename in os.listdir("/content/SOCOFing/Altered/Altered-Hard"):
        data.append(filename)
        data.append("altered")
        writer.writerow(data)
        data=[]
    for filename in os.listdir("/content/SOCOFing/Altered/Altered-Medium"):
        data.append(filename)
        data.append("altered")
        writer.writerow(data)
        data=[]
writeFile.close()
```
